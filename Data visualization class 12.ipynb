```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
euro = pd.read_csv("euro.csv")
```

```python
euro
```

```python
euro.columns
```

```python
euro_dollar = euro[[r"Period\Unit:", r"[US dollar ]"]].copy()
```

```python
euro_dollar
```

```python
euro_dollar.rename(columns = {r"Period\Unit:":"Date", r"[US dollar ]": "USD"}, inplace = True)
```

```python
euro_dollar
```

```python
euro_dollar.Date = pd.to_datetime(euro_dollar.Date)
```

```python
euro_dollar
```

```python
euro_dollar.sort_values(by = "Date", inplace = True)
```

```python
euro_dollar
```

```python
euro_dollar.reset_index(drop = True, inplace = True)
```

```python
euro_dollar
```

```python
euro_dollar.info()
```

```python
euro_dollar["USD"].value_counts()
```

```python
bol = euro_dollar["USD"] != "-"
```

```python
euro_dollar = euro_dollar[bol]
```

```python
euro_dollar.USD = euro_dollar.USD.astype(float)
```

```python
import matplotlib.style as style
```

```python
style.use("fivethirtyeight")
```

```python
plt.plot(euro_dollar["Date"], euro_dollar.USD)
plt.show()
```

```python
Moving Windows

* 7 Days
* 30 Days
* 50 Days
* 100 Days
* 365 Days
```

```python
euro_dollar.USD.rolling(7).mean().head(50)
```

```python
rolling = [7,30,50,100,365]
plots = [2,3,4,5,6]
```

```python
plt.figure(figsize = (11,9))
plt.subplot(3,2,1)
plt.plot(euro_dollar["Date"], euro_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3,2,plot)
    plt.plot(euro_dollar["Date"], euro_dollar.USD.rolling(rol).mean(), linewidth = 3)
    plt.title(f"Moving Average {rol} days")
    
plt.tight_layout()
plt.show()
```

```python
euro_dollar["Rolling_mean_30"] = euro_dollar.USD.rolling(30).mean()
```

```python
euro_dollar
```

```python
bol_2006_2009 = (euro_dollar.Date.dt.year >= 2006) & (euro_dollar.Date.dt.year <= 2009)
```

```python
bol_2007_2008 = (euro_dollar.Date.dt.year >= 2007) & (euro_dollar.Date.dt.year <= 2008)
```

```python
financial_crisis = euro_dollar[bol_2006_2009].copy()
financial_crisis
```

```python
financial_crisis_07_08 = financial_crisis[bol_2007_2008].copy()
```

```python
fig, ax = plt.subplots(figsize = (8,3))
ax.plot(financial_crisis["Date"], financial_crisis.Rolling_mean_30, linewidth = 1)
ax.plot(financial_crisis_07_08["Date"], financial_crisis_07_08.Rolling_mean_30, linewidth = 3, color = "#800080")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,3))
ax.plot(financial_crisis["Date"], financial_crisis.Rolling_mean_30, linewidth = 1)
ax.plot(financial_crisis_07_08["Date"], financial_crisis_07_08.Rolling_mean_30, linewidth = 3, color = "#800080")

ax.set_xticklabels([])
ax.set_yticklabels([])
plt.show()
```
