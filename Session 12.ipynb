```python
euro_dollar.Date = pd.to_datetime(euro_dollar.Date)
```

```python
euro_dollar
```

```python
Moving Windows

* 7 Days
* 30 Days
* 50 Days
* 100 Days
* 365 Days
```

```python
euro_dollar.USD.rolling(7).mean().head(50)
```

```python
rolling = [7,30,50,100,365]
plots = [2,3,4,5,6]
```

```python
plt.figure(figsize = (11,9))
plt.subplot(3,2,1)
plt.plot(euro_dollar["Date"], euro_dollar.USD)
plt.title("Original")

for rol, plot in zip(rolling, plots):
    plt.subplot(3,2,plot)
    plt.plot(euro_dollar["Date"], euro_dollar.USD.rolling(rol).mean(), linewidth = 3)
    plt.title(f"Moving Average {rol}")
    
plt.tight_layout()
plt.show()
```

```python
euro_dollar["Rolling_mean_30"] = euro_dollar.USD.rolling(30).mean()
```
